```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.mlab as mlab
import seaborn as sns
```

```python
df = pd.read_excel('/content/DMM Experimental data for ML.xlsx')
df.head()
```

```python
df.columns
```

```python
plt.figure(figsize=(10, 6))
ax = sns.boxplot(x='Water_Cement Ratio', y='Avg_UCS', data=df)
```

```python
plt.figure(figsize=(10, 6))
ax = sns.stripplot(x='Water_Cement Ratio', y='Avg_UCS', data=df)
```

```python
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='Water_Cement Ratio', y='Avg_UCS', data=df)
```

```python
plt.figure(figsize=(10, 6))
ax = sns.stripplot(x='Cement contents', y='Avg_UCS', data=df)
```

```python
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='Cement contents', y='Avg_UCS', data=df)
```

```python
plt.figure(figsize=(10, 6))
ax = sns.violinplot(x='Cement contents', y='Avg_UCS', data=df)
```

```python
plt.figure(figsize=(10, 6))
ax = sns.pointplot(x='Cement contents', y='Avg_UCS', data=df)
```

```python
plt.figure(figsize=(10, 6))
ax = sns.lineplot(x='Cement contents', y='Avg_UCS', data=df)
```

```python
plt.figure(figsize=(10, 8))
ax = sns.jointplot(x='Cement contents', y='Avg_UCS', data=df)
```

```python
plt.figure(figsize=(10, 6))
ax = sns.regplot(x='Cement contents', y='Avg_UCS', data=df)
```

```python
plt.figure(figsize=(10, 6))
ax = sns.catplot(x='Cement contents', y='Avg_UCS', data=df)
```
